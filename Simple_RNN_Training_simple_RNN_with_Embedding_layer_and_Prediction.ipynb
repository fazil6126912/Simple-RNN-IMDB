#Word Embedding

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## sentences
sent=[
'the glass of milk',
'the glass of juice',
'the cup of tea',
'I am a good boy' ,
'I am a good developer',
'understand the meaning of words' ,
'your videos are good',]
```

```python
sent
```

```python
vocb_size = 10000
sent = [one_hot(word,vocb_size) for word in sent]
sent
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length = 8
embedded_docs = pad_sequences(sent,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
dim = 10
model = Sequential()
model.add(Embedding(vocb_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```

#Preprocessing IMDB dataset for RNN

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
max_features = 10000
maxlen = 500
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
X_train = sequence.pad_sequences(X_train, maxlen=maxlen)
X_test = sequence.pad_sequences(X_test, maxlen=maxlen)
```

```python
X_train[0]
```

#Training simple RNN with Embedding layer

```python
model = Sequential()
model.add(Embedding(max_features, 128,input_length = maxlen))
model.add(SimpleRNN(128,activation = 'relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stopping]
)
```

```python
model.save('simple_rnn_model.h5')
```

#Prediction

```python
word_index = imdb.get_word_index()
reverse_word_index = {value:key for key,value in word_index.items()}
```

```python
def decode_review(text):
    return ' '.join([reverse_word_index.get(i-3,'?') for i in text])

def preprocess_text(text):
  words = text.lower().split()
  encoded_review = [word_index.get(word,2) + 3 for word in words]
  padded_review = sequence.pad_sequences([encoded_review], maxlen=maxlen)
  return padded_review
```

```python
def predict_sentiment(text):
  padded_review = preprocess_text(text)
  prediction = model.predict(padded_review)[0][0]
  sentiment = 'positive' if prediction > 0.5 else 'negative'
  return sentiment, prediction
```

```python
text = "A visual treat with a strong story & characters A unique genre, a well written story (script) & strong characters give this movie a strong start."
sentiment, prediction = predict_sentiment(text)
sentiment, prediction
```
